```python
import pandas as pd

url = 'https://drive.google.com/uc?id=1kd0HDnchsnitXhgcMymtSBTl9gRe8SEj&export=download'
nba_teams = pd.read_csv(url)
```

```python
nba_teams
```

```python
#reads teams playing based on input
away = input("What is the away team?")
home = input("What is the home team?")
```

```python
#prints matchup
print(f'The matchup is {away} vs. {home}')
```

```python
def select_player_against_worst(away, home, nba_teams):
    # Find the team rows for Team A and Team B
    away_nba_teams = nba_teams[nba_teams['Unnamed: 0'] == away]
    home_nba_teams = nba_teams[nba_teams['Unnamed: 0'] == home]

    if away_nba_teams.empty or home_nba_teams.empty:
        return "One of the teams specified does not exist in the dataset."

    # Get the worst position for both teams
    worst_position_home = home_nba_teams['Worst'].values[0]
    worst_position_away = away_nba_teams['Worst'].values[0]

    # Find the player that plays against the worst position
    player_against_worst = away_nba_teams[worst_position_home].values[0]
    player_against_worst1 = home_nba_teams[worst_position_away].values[0]

    #determine statistic to bet on depending on player position and who they are against
    if worst_position_away == worst_position_home:
        bet_away = f"{player_against_worst} Under Assists"
        bet_home = f"{player_against_worst1} Under Assists"
    else:
        if worst_position_away == 'C':
            bet_away = f"{player_against_worst} Over Rebounds"
        elif worst_position_away == 'SG':
            bet_away = f"{player_against_worst} Over Points"
        else:
            bet_away = f"{player_against_worst} Over Assists"

        if worst_position_home == 'C':
            bet_home = f"{player_against_worst1} Over Rebounds"
        elif worst_position_home == 'SG':
            bet_home = f"{player_against_worst1} Over Points"
        else:
            bet_home = f"{player_against_worst1} Over Assists"

    #return both bets
    return f"{bet_away}. {bet_home}"

#print bets
result = select_player_against_worst(away, home, nba_teams)
print(result)
```
